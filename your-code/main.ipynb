#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Import users table:

```python
users = pd.read_csv('users_table.csv')
```

#### 3. Rename Id column to userId

```python
users.rename(columns = {'Id':'userId'}, inplace = True)
users
```

#### 4. Import posts table:

```python
posts = pd.read_csv('posts_table.csv')
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
posts.rename(columns = {'Id':'postId','OwnerUserId':'userId'}, inplace = True)
posts.head(0)
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
users_f = users[['userId','Reputation','Views','UpVotes','DownVotes']]
posts_f = posts[['postId', 'Score', 'userId','ViewCount','CommentCount']]
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
#merged = users_f.append(posts_f)
frames = [users_f, posts_f]
  
merged = pd.concat(frames)
merged
```

#### 8. How many missing values do you have in your merged dataframe? On which columns?

```python
merged.isna().sum().sum()
for column in merged:
    a = merged[column].isna().sum()
    if a > 0:
        print('column '+ column + ' has ' + str(a) + ' Nan values')
```

#### 9. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
merged = merged.fillna(0)
#there is important info in all the columns the 2 df are completing each other. Even though having a lot of Nan I should not delete them 
merged.head(2)
```

#### 10. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
#we should change it to integers. All but the Score
merged["userId"] = merged['userId'].astype(int)
merged["Reputation"] = merged['Reputation'].astype(int)
merged["Views"] = merged['Views'].astype(int)
merged["UpVotes"] = merged['UpVotes'].astype(int)
merged["DownVotes"] = merged['DownVotes'].astype(int)
merged["postId"] = merged['postId'].astype(int)
merged["ViewCount"] = merged['ViewCount'].astype(int)
merged["CommentCount"] = merged['CommentCount'].astype(int)

merged
```
